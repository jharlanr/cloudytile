```python

# 1. Directory containing your .nc files
nc_dir = "/home/jupyter/data/2019cw_tstacks"
# 2. Directory where you want to save PNGs
out_dir = "/home/jupyter/data/2019cw_pngs"
os.makedirs(out_dir, exist_ok=True)

# 3. The name of the variable inside the NetCDF you want to export
var_name = "reflectance"  # e.g. "B04" or "band_data"

for nc_path in glob.glob(os.path.join(nc_dir, "*.nc")):
    ds = xr.open_dataset(nc_path)
    da = ds[var_name]              # dims: (time, band, y, x)
    da = da.isel(band=[0,1,2])     # pick bands 0,1,2 as R,G,B
    
    for ti in range(len(da.time)):
        frame3 = da.isel(time=ti)           # dims: (band=3, y, x)
        arr3 = frame3.values/10000                # shape (3, Y, X)
        
        # transpose → (Y, X, 3)
        rgb = np.transpose(arr3, (1, 2, 0))
        rgb = np.clip(rgb, 0.0, 1.0)
        
        fname = f"{os.path.splitext(os.path.basename(nc_path))[0]}_t{ti:03d}.png"
        out_path = os.path.join(out_dir, fname)
        plt.imsave(out_path, rgb)  # full-color PNG
        
    ds.close()
    
```
